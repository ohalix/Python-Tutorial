```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
```

```python
df_us
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')

df_us.head()
```

```python
job_month_df = df_us.pivot_table(index= 'job_posted_month', columns= 'job_title_short', aggfunc= 'size')

job_month_df.reset_index(inplace= True)
job_month_df['month_no'] = pd.to_datetime(job_month_df['job_posted_month'], format= '%B').dt.month

job_month_df['month_no']
```

```python
job_month_df.sort_values('month_no', inplace= True)
job_month_df.set_index('job_posted_month', inplace= True)
job_month_df.drop(columns= 'month_no', inplace= True)


job_month_df
```

```python
top_3 = df_us['job_title_short'].value_counts().head(3)
top_3 = top_3.index


job_month_df[top_3].plot(kind= 'line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
